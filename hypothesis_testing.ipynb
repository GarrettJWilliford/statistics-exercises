```python

def pt_score(x1, x2, n1, n2):
    xbar1 = x1.mean()
    xbar2 = x2.mean()
    s1 = x1.std()
    s2 = x2.std()
    degf = n1 + n2
    s_p = sqrt(((n1 - 1) * s1**2 + (n2 - 1) * s2**2) / (n1 + n2 - 2))
    standard_error = se = sqrt(s1**2 / n1 + s2**2 / n2)
    t = (xbar1 - xbar2) / (s_p * sqrt(1/n1 + 1/n2))
    p = stats.t(degf).cdf(t) * 2
    return t, p



t, p = pt_score(stats.norm(90, 15), stats.norm(100, 20), 40, 50)
print(t)
p
```
